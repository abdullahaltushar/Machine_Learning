```python
from sklearn.datasets import load_digits
import pandas as pd
import seaborn as sns

dataset = load_digits()
dataset.keys()
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
plt.gray()
plt.matshow(dataset.data[0].reshape(8,8))
```

```python
pd.DataFrame(dataset.data)
```

```python
df=pd.DataFrame(dataset.data, columns=dataset.feature_names)
df.head()
```

```python
df.describe()
```

```python
x=df
y=dataset.target
```

```python
y.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(x)
```

```python
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,Y_train, Y_test=train_test_split(X_scaled,y,test_size=0.2, random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,Y_train)
model.score(X_test,Y_test)
```

```python
from sklearn.decomposition import PCA
pca=PCA(0.95)
X_pca= pca.fit_transform(x)
X_pca.shape
```

```python
X_pca
```

```python
pca.explained_variance_ratio_
```

```python
pca.n_components_
```

```python
X_train_pca,X_test_pca, Y_train_pca,Y_test_pca=train_test_split(X_pca,y, test_size=0.2,random_state=30)
```

```python
model=LogisticRegression(max_iter=1000)
model.fit(X_train_pca,Y_train_pca)
model.score(X_test_pca,Y_test_pca)
```

```python
from sklearn.datasets import load_breast_cancer
cencer=load_breast_cancer()
```

```python
cencer.keys()
```

```python
cencer.data
```

```python
cencer.data.shape
```

```python
cencer.data[0]
```

```python
df1=pd.DataFrame(cencer.data, columns=cencer.feature_names)
df1.head()
```

```python
df1.describe()
```

```python
x1=df1
y1=cencer.target
```

```python
x1.shape
```

```python
y1.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaled=scaler.fit_transform(x1)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
x_pca= pca.fit_transform(X_scaled)
x_pca.shape
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=cencer['target'])
```

```python
x_train_pca,x_test_pca, y_train_pca,y_test_pca=train_test_split(x_pca,y1, test_size=0.2,random_state=30)
```

```python
model=LogisticRegression(max_iter=1000)
model.fit(x_train_pca,y_train_pca)
model.score(x_test_pca,y_test_pca)
```
