```python
# seaborn is a statistical plotting library
# It has beautiful default styles
# It also is designed to work very well with pandas dataframe objects 
```

```python
 import seaborn as sns
```

```python
%matplotlib inline
```

```python
#load sns dataset
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# change bin vlue foe specific plot
sns.displot(tips['total_bill'],bins=40)
```

```python
#kind value change plot hexagone
sns.jointplot(x='total_bill',y='tip', data=tips,kind='hex')
```

```python
#kind value change plot regression 
sns.jointplot(x='total_bill',y='tip', data=tips,kind='reg')
```

```python
#kind value change plot 2D
sns.jointplot(x='total_bill',y='tip', data=tips,kind='kde')
```

```python
#details all column full data set
sns.pairplot(tips)
```

```python
#sex ways details
sns.pairplot(tips,hue='sex')
```

```python
#how many bins result is allocate with result
sns.rugplot(tips['total_bill'])
```

```python
import numpy as np
```

```python
#total male and female
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
#countplot
sns.countplot(x='sex',data=tips)
```

```python
#box plot use day
sns.boxplot(x='day',y='total_bill', data=tips)
```

```python
#box plot use day and category smoker 
sns.boxplot(x='day',y='total_bill', data=tips,hue='smoker')
```

```python
#violineplot
sns.violinplot(x='day',y='total_bill', data=tips)
```

```python
sns.violinplot(x='day',y='total_bill', data=tips,hue='sex')
```

```python
sns.stripplot(x='day',y='total_bill', data=tips,jitter=True, hue='sex',split=True)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips,color='black')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='violin')
```

# Matrix plots

```python
flights=sns.load_dataset('flights')
flights.head()
```

```python
tc=tips.corr()
```

```python
sns.heatmap(tc,annot=True)
```

```python
fp=flights.pivot_table(index='month',columns='year', values='passengers')
```

```python
sns.heatmap(fp)
```

```python
sns.heatmap(fp,cmap='coolwarm', linecolor='black',linewidths=3)
```

```python
sns.clustermap(fp)
```

```python
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
```

# Grids 

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
import matplotlib.pyplot as plt
g=sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g=sns.PairGrid(iris)
g.map_diag(sns.displot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g=sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.displot,'total_bill')
```

```python
g=sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(plt.scatter,'total_bill','tip')
```

# Regression Plots

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex',markers=['o','v'],
          scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex', row='time')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', hue='sex', aspect=0.6, size=8)
```
