```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
##create x and y value
## linsepace(1st value, endvalue, totol value)
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
#functional method
#plot(Xvalue, yvalue, color)
plt.plot(x,y,'r-')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

#subplot(row,column, plotnumber)
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#object orianted method

fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set
```

```python
fig=plt.figure()
#add_axes(left,up,x_asix_height, y_axis height)
axes1=fig.add_axes([0.1,0.1,0.8,0.8])

axes2=fig.add_axes([0.2,0.4,0.3,0.3])
#plot show axes1
axes1.plot(x,y)
axes1.set_title("larger plot")
#plot show axes2
axes2.plot(y,x)
axes2.set_title("smaller plot")
```

```python
fig=plt.figure()
#add_axes(left,up,x_asix_height, y_axis height)
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y)
```

subplot

```python
# subplots(row, col)
fig,axes=plt.subplots(nrows=1,ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
# subplots(row, col)
fig,axes=plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('first plot')
axes[1].plot(y,x)
axes[1].set_title('Second plot')
```

figure size and dpi

```python
fig=plt.figure(figsize=(3,2))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,8))
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
#save figure
fig.savefig('my_picture.png',dpi=200)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('Title')
ax.set_ylabel('Y')
ax.set_xlabel('x')
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2)
ax.plot(x,x**3)
```

```python
#x.legend foe label
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3, label='Y cubed')
#legend(location you generate use loc=(x,y))
ax.legend(loc=(0.1,0.5))
```

plot appearance

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
# add plot color and linewidth(lw) and alpha(color display) and line style(ls) and marker and marker size
ax.plot(x,y,color='orange',linewidth=4,alpha=0.5,lineStyle="--", marker="o", markersize=20, 
        markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='black') # use color code use google 
```

```python
# generate x and y size
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='orange',linewidth=2,ls='--')
## set_xlim( lower bound, upper bound)
ax.set_xlim([0,1])
ax.set_ylim([0,3])
```
