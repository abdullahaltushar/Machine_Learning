```python
#import pandas
import pandas as pd
```

```python
#read file
df=pd.read_csv("dc.csv")
df
```

```python
#separate input data
input=df.drop('salary more than 100k',axis='columns')
input
```

```python
#separate target data
target=df['salary more than 100k']
target
```

# use labelEncoder convet test to digits

```python
#import labelEncoder
from sklearn.preprocessing import LabelEncoder
le_company=LabelEncoder()
le_job=LabelEncoder()
le_degree=LabelEncoder()
```

```python
#company data encode
input['company_data']=le_company.fit_transform(input['company'])
input
```

```python
#job data encode
input['job_data']=le_company.fit_transform(input['job'])
input
```

```python
#degree data encode
input['degree_data']=le_company.fit_transform(input['degree'])
input
```

```python
#delete company job and degree data
input.drop(['company','job','degree'],axis='columns',inplace=True)
input
```

```python
#import model
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
#model train
model.fit(input,target)
```

```python
#model score
model.score(input,target)
```

Is salary of BSRM,it, master degree > 100 k ?

```python
#predict data
model.predict([[1,1,1]])
```
